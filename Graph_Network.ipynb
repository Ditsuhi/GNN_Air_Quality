```python
# Import system modules
import arcpy
from itertools import zip_longest

# Set local variables
arcpy.env.workspace = r"C:\Users\ditsuhi\Documents\ArcGIS\Projects\metStation\metStation.gdb"
input_table = r"C:\Users\ditsuhi\Documents\ArcGIS\Projects\metStation\metStation.gdb\cent_XYTableToPointFINISH"

arcpy.AddField_management(input_table, "latitude1", "DOUBLE")
arcpy.AddField_management(input_table, "longitude1", "DOUBLE")



fields = ["latitude","latitude1", "longitude", "longitude1"]


with arcpy.da.SearchCursor(input_table, fields) as n1_cur:
    with arcpy.da.UpdateCursor(input_table, fields) as n_cur:
        filler_row = next(n1_cur)
        
        for n_row, n1_row in zip_longest(n_cur, n1_cur, fillvalue=filler_row):            
            n_row[1] = n1_row[0]
            n_row[3] = n1_row[2]
            n_cur.updateRow(n_row)


out_lines = "cent_PointLine"


#for i in range(24):
arcpy.XYToLine_management(input_table, out_lines, "latitude", "longitude",  
                        "latitude1", "longitude1", "GEODESIC", attributes="ATTRIBUTES")
```
